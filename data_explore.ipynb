<h2 align='center'>Searching for patterns</h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gabrielalastra/Supermarkt/main/WizardSupermarkt_database.csv')
```

```python
df.head()
```

<h2>Search for standart</h2>

```python
plt.figure(figsize=[7,5])
sns.heatmap(df.corr())
plt.title('\nNumerical Columns Corelation\n')
```

```python
sns.pairplot(df, hue='Sales', palette='coolwarm')
```

```python
features=['Ship Mode', 'State', 'Category', 'Sub-Category']
for i in features:
  plt.figure(figsize=[30,10])
  g = sns.countplot(x=i, data=df)
```

```python
plt.figure(figsize=[30,10])
sns.countplot(x='State', data=df, hue='Ship Mode')
plt.title('\nShip Mode per State\n', size=20)
```

```python
df['Sales Agent ID'].isna().value_counts()
```

```python
df['Sales Agent ID']=df['Sales Agent ID'].fillna(0)
```

```python
df['Sales'].isna().value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = ['Sales']
df[scale] = MinMaxScaler().fit_transform(df[scale])
```

```python
plt.figure(figsize=[60,10])
plt.title('\nQuantity of Products by Sales Order\n',size=25)
sns.countplot(x='City', data=df, hue='Quantity')
```
